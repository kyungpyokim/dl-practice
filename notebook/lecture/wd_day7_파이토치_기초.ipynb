## gradient 계산 허용/비허용

tensor의 requires_grad는 자동미분기능 활성화

### 역전파 전

```python
t = torch.tensor([1.0], requires_grad=True)
a = t + 1
b = a * 2
c = b**2

print('t =', t)
print('a =', a)
print('b =', b)
print('c =', c)
```

```python
print('t.data =', t.data)
print('t =', t.grad)
print('a =', a.grad)
print('b =', b.grad)
print('c =', c.grad)
```

### 역전파 후

```python
t = torch.tensor([1.0], requires_grad=True)
a = t + 1
b = a * 2
c = b**2
c.backward()

print('t =', t)
print('a =', a)
print('b =', b)
print('c =', c)
print(t.grad)
```

```python
print('t.data =', t.data)
print('t =', t.grad)
print('a =', a.grad)
print('b =', b.grad)
print('c =', c.grad)
```

### retain_grad 로 저장

```python
t = torch.tensor([1.0], requires_grad=True)
a = t + 1
b = a * 2
c = b**2

a.retain_grad()
b.retain_grad()
c.retain_grad()

c.backward(retain_graph=True)

print('t =', t)
print('a =', a)
print('b =', b)
print('c =', c)
print(t.grad)
```

```python
print('t.data =', t.data)
print('t =', t.grad)
print('a =', a.grad)
print('b =', b.grad)
print('c =', c.grad)
```

역전파 한 번 더 하면

```python
c.backward(retain_graph=True)

print('t =', t)
print('a =', a)
print('b =', b)
print('c =', c)
print(t.grad)
```

```python
print('t.data =', t.data)
print('t =', t.grad)
print('a =', a.grad)
print('b =', b.grad)
print('c =', c.grad)
```

### grad 초기화

```python
t.grad.zero_()
a.grad.zero_()
b.grad.zero_()
c.grad.zero_()

print('t =', t.grad)
print('a =', a.grad)
print('b =', b.grad)
print('c =', c.grad)
```

# 인공신경망 만들기

```python
import torch.nn as nn
import torch.optim as optim
```

## 인공신경망

### 기본 틀

```python
class Net(nn.Module):  # 클래스 정의 및 상속
    def __init__(self):  # 신경망 레이어를 정의하기 위한 생성자
        super(Net, self).__init__()  # 신경망 초기화
        # ...
        # ...

    def forward(self, x):  # 순전파 연산 정의
        # ...
        # ...
        return x
```

### 레이어 추가 예시

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(in_features=5, out_features=3, bias=True)
        self.fc1_act = nn.ReLU()
        self.fc2 = nn.Linear(in_features=3, out_features=1, bias=True)

    def forward(self, x):
        out = self.fc1(x)
        out = self.fc1_act(out)
        out = self.fc2(out)
        return out
```

```python
model = Net()
model
```

## 파라미터 정보 알아보기

파라미터 정보

```python
params = list(model.parameters())
print(params)
```

크기

```python
print(len(params))

for i in range(len(params)):
    print(params[i].size())
```

각 파라미터의 이름

```python
for name, param in model.named_parameters():
    print(name)
```

```python
batch_size = 1
input = torch.randn(batch_size, 5)
output = model(input)

target = torch.rand(batch_size, 3)
criterion = nn.MSELoss()

loss = criterion(output, target)
print(loss)
```
