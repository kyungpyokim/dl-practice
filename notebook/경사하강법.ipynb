```python
import matplotlib.pyplot as plt
import numpy as np

X = np.random.rand(100)
y = 0.2 * X
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(X, y)
```

```python
w = np.random.uniform(-1, 1)
b = np.random.uniform(-1, 1)

pred = w * X + b


def plot_pred(pred, y):
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y)
    plt.scatter(X, pred)
    plt.show()


plot_pred(pred, y)
```

```python
lr = 0.5  # learning rate

for epoch in range(300):
    w_grad = 0  #   
    b_grad = 0

    y_pred = w * X + b  #  

    loss = np.abs(y_pred - y).mean()  # 

    w_grad = lr * ((y_pred - y) * X).mean()
    b_grad = lr * (y_pred - y).mean()
    #  

    #  :  
    w = w - w_grad  #  
    b = b - b_grad

    if loss < 0.001:
        break

    print("Loss : ", loss)
    if epoch % 10 == 0:
        plot_pred(y_pred, y)
```
